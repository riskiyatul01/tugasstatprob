RANGE

```python
# Fungsi untuk menghitung rentang
def calculate_range(jumlahgol):
    nilai_tertinggi = max(jumlahgol)
    print(f"Nilai tertinggi: {nilai_tertinggi}")

    nilai_terendah = min(jumlahgol)
    print(f"Nilai terendah: {nilai_terendah}")

    # Selisih antara nilai tertinggi dan terendah
    data_range = max(jumlahgol) - min(jumlahgol)
    
    print(f"\n")    
    print(f"Range: {nilai_tertinggi} - {nilai_terendah} = {data_range}")  
    
    return data_range


# Inisialisasi dataset
jumlahgol = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
print(f"jumlah gol: {jumlahgol}")

data_range = calculate_range(jumlahgol)
```

```python
import matplotlib.pyplot as plt

# Create a histogram
plt.hist(jumlahgol, bins=30, edgecolor='black', alpha=0.5, color='green')

# Define the horizontal line coordinates
y = 1.5 # The y-coordinate where the line will be drawn
x1 = 4/70*1  # The x-coordinate of the start point
x2 = 66/70*1  # The x-coordinate of the end point

# Add a horizontal line
plt.axhline(y=y, xmin=x1, xmax=x2, color='red', linestyle='dashed', linewidth=1, label=f'Range = {calculate_range(jumlahgol)}')

# Add vertical line
plt.axvline(x=min(jumlahgol), color='blue', linestyle='dashed', linewidth=1, label=f'Min Data: {min(jumlahgol):.2f}')
plt.axvline(x=max(jumlahgol), color='blue', linestyle='dashed', linewidth=1, label=f'Max Data: {max(jumlahgol):.2f}')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Range')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

VARIANCE

```python
# Fungsi untuk menghitung rata-rata
def calculate_mean(jumlahgol):
    return sum(jumlahgol) / len(jumlahgol)

# Fungsi untuk menghitung variansi
def calculate_variance(jumlahgol):
    # Rata-rata 
    mean = calculate_mean(jumlahgol)
    print(f"Rata-rata: {mean}")  
    
    # Kuadrat selisih antara setiap nilai data dengan rata-rata
    print(f"Kuadrat selisih setiap nilai data dengan rata-rata:")    
    squared_diff = []
    for x in jumlahgol:
        squared_diff_per_x = (x - mean)**2
        squared_diff.append( squared_diff_per_x )
        print(f"({x} - {mean})**2 = {(x - mean)}**2 = {squared_diff_per_x}")    
    
    print(f"Total kuadrat selisih: {sum(squared_diff)}")    
    print(f"Jumlah total data: {len(jumlahgol)}")    
    
    # Jumlah kuadrat selisih antara setiap nilai data dengan rata-rata, kemudian dibagi dengan jumlah total data
    variance = sum(squared_diff) / len(jumlahgol)
    
    print(f"\n")    
    print(f"Variansi: {sum(squared_diff)} / {len(jumlahgol)} = {variance}")    
    
    return variance


# Inisialisasi dataset
data = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
print(f"Data: {jumlahgol}")

variance = calculate_variance(jumlahgol)
```

STANDAR DAVIATION

```python
# Fungsi untuk menghitung standar deviasi
def calculate_standard_deviation(jumlahgol):
    variance = calculate_variance(jumlahgol)
    standard_deviation = variance**0.5
    print(f"\n")   
    print(f"Standard Deviation: {variance}**0.5 = {standard_deviation}")
    
    return standard_deviation

# Inisialisasi dataset
jumlahgol = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
print(f"jumlah gol: {jumlahgol}")

standard_deviation = calculate_standard_deviation(jumlahgol)
mean = calculate_mean(jumlahgol)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Buat histogram
plt.hist(jumlahgol, bins=30, edgecolor='black', alpha=0.5, color='green')

# Tambahkan garis vertikal pada standar deviasi
plt.axvline(x=mean, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean}')
plt.axvline(x=mean - standard_deviation, color='blue', linestyle='dashed', linewidth=1, label=f'{mean-standard_deviation:.2f}')
plt.axvline(x=mean + standard_deviation, color='blue', linestyle='dashed', linewidth=1, label=f'{mean+standard_deviation:.2f}')

# Tambahkan label dan judul
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title(f'Std Deviation = {standard_deviation:.2f}')

# Tambahkan legenda
plt.legend()

# Tampilkan plot
plt.show()
```
